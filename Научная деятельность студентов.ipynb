```python
def nan_table(dat):
        nan = dat.isnull().sum()
        nan_percent = 100 * dat.isnull().sum() / len(dat)
        nan_table = pd.concat([nan, nan_percent], axis=1)
        nan_table_new = nan_table.rename(
        columns = {0 : 'Пустые ячейки', 1 : '% от всех ячеек столбца'})
        nan_table_new = nan_table_new[
            nan_table_new.iloc[:,1] != 0].sort_values(
        '% от всех ячеек столбца', ascending=False).round(1)
        return nan_table_new
```

```python
nan_table(dat)
```

```python
d = nan_table(dat)
to_delete = list(d[d['% от всех ячеек столбца'] >= 99.0].index)
dat1 = dat.drop(columns = list(to_delete))
print(dat1)
```

```python
import pandas as pd
import numpy as np
import matplotlib
import seaborn as sns 
from collections import Counter 
from collections import OrderedDict

dat = pd.read_excel('МСЖ_2018.xlsx')

def nan_table(dat):
        nan = dat.isnull().sum()
        nan_percent = 100 * dat.isnull().sum() / len(dat)
        nan_table = pd.concat([nan, nan_percent], axis=1)
        nan_table_new = nan_table.rename(
        columns = {0 : 'Пустые ячейки', 1 : '% от всех ячеек столбца'})
        nan_table_new = nan_table_new[
            nan_table_new.iloc[:,1] != 0].sort_values(
        '% от всех ячеек столбца', ascending=False).round(1)
        return nan_table_new
    
d = nan_table(dat)
to_delete = list(d[d['% от всех ячеек столбца'] >= 99.0].index)
dat1 = dat.drop(columns = list(to_delete))

dat1.loc[dat1['сводный индекс по науке'] == 'вовлечен в научно-исследовательскую деятельность', 'сводный индекс по науке'] = 1
dat1.loc[dat1['сводный индекс по науке'] == 'не вовлечен в научно-исследовательскую деятельность', 'сводный индекс по науке'] = 0

sci_data = dat1[dat1['сводный индекс по науке'] == 1] 

cor_dat1 = dat1.corr()['сводный индекс по науке'].sort_values(ascending = False) #нахождение корреляции

sci_datum = sci_data[sci_data['Желаемая область работы после университетета'] == 'В образовательных и научно-исследовательских учреждениях']

#примеры графиков и диаграмм

fac_sci = sci_datum['факультет'].value_counts()
fac_all = dat1['факультет'].value_counts()
percent = (fac_sci*100)/fac_all
p = percent.dropna().sort_values(ascending = False)
plt.figure(figsize=(8,6))
sns.barplot(y=p.index, x=p, palette = 'muted')
plt.title('Число студентов вовлечённых в науку по факультетам, %')
plt.xlabel('%')
plt.savefig('fac.png')
plt.show()
####
all_olymp = sci_datum[sci_datum['Принимал(а) участие во всероссийской олимпиаде для студентов "Я – профессионал"'] == 1]
all_olymp = all_olymp[all_olymp['Принимал(а) участие в других всероссийских студенческих олимпиадах и конкурсах'] == 1]
no_olymp = sci_datum[sci_datum['Принимал(а) участие во всероссийской олимпиаде для студентов "Я – профессионал"'] == 0]
no_olymp = no_olymp[no_olymp['Принимал(а) участие в других всероссийских студенческих олимпиадах и конкурсах'] == 0]
prof = sci_datum[sci_datum['Принимал(а) участие во всероссийской олимпиаде для студентов "Я – профессионал"'] == 1]
prof = prof[prof['Принимал(а) участие в других всероссийских студенческих олимпиадах и конкурсах'] == 0]
other = sci_datum[sci_datum['Принимал(а) участие в других всероссийских студенческих олимпиадах и конкурсах'] == 1]
other = other[other['Принимал(а) участие во всероссийской олимпиаде для студентов "Я – профессионал"'] == 0]

all_olymp = int(all_olymp['Принимал(а) участие во всероссийской олимпиаде для студентов "Я – профессионал"'].value_counts())
no_olymp = int(no_olymp['Принимал(а) участие во всероссийской олимпиаде для студентов "Я – профессионал"'].value_counts())
prof = int(prof['Принимал(а) участие во всероссийской олимпиаде для студентов "Я – профессионал"'].value_counts())
other = int(other['Принимал(а) участие во всероссийской олимпиаде для студентов "Я – профессионал"'].value_counts())

plt.figure(figsize=(8,6))
labels = 'Участвовал во всех упомянутых олимпиадах', 'Не участвовал в олимпиадах', 'Участвовал только в олимпиаде "Я - профессионал"', 'Участвовал в других олимпиадах и конкурсах'
sizes = [all_olymp, no_olymp, prof, other]
colors = ['green', 'blue', 'red', 'purple']

plt.pie(sizes, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=90)

plt.axis('equal')
plt.title('Участие студентов в олимпиадах')
plt.show()
####
all_mark = dat1['Оценка за КР/ВКР'].value_counts()
sci_mark = sci_datum['Оценка за КР/ВКР'].value_counts()
plt.figure(figsize=(8,6))
sns.lineplot(x=all_mark.index, y = all_mark, color='blue')
sns.lineplot(x=sci_mark.index, y = sci_mark, color='purple')
plt.title('Оценка студентов за КР/ВКР')
plt.xlabel('балл')
plt.ylabel('человек')
plt.savefig('KrAndVKR.png')
plt.show()
####
plt.figure(figsize=(18,10))
sns.countplot(y='курс+уровень образования для кросстабов',data = sci_datum, hue = 'Опубликованные научные работы - у меня нет публикаций', palette='coolwarm')
plt.ylabel('курс')
plt.show()
#подсчёт комментариев касательно СМИ (блоги и каналы аналогично)
from pandas import DataFrame

smi_com = list(sci_datum['Комментарий к СМИ за последние 3 месяца'].dropna())
spis = []
for item in smi_com:
    i = item.split(',')
    for k in i:
        new = k.strip()
        spis.append(new)
        new = sorted(spis)

df = DataFrame({'СМИ': new})
df.to_excel('SMI.xlsx', sheet_name='sheet1', index=False)
#запишем получившийся список в файл excel и почистим его с помощью OpenRefine.
df_new = pd.read_excel('SMI1.xlsx')
SMI = Counter(df_new['СМИ'])
s = dict(OrderedDict(sorted(SMI.items(), key=lambda t: t[1]))) #готовый словарь с подсчитанными значениями
```
